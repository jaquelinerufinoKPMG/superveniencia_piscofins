```python
processor = TelasPretasProcessor(
    pasta_entrada=r"input\anexo_a",
    pasta_output=r"output\anexo_a",
    usar_prefixo_da_subpasta=False,
)

# 1) Processar todos os txts da árvore
processor.processar_txts_da_arvore()

# 2) Processar listagem específica
processor.processar_listagem(
    arquivo_entrada=r"input\anexo_a\ListagemCadastroConsultaBens.txt",
    pasta_saida=r"output\anexo_a\ListagemCadastroConsultaBens",
)

# 3) Validar gerados e gerar Excel
processor.validar_gerados(
    txt_contratos_esperados=r"data\contratos2014.txt",
    raiz_gerados=r"output\anexo_a",
    pasta_saida_validacao=r"output\anexo_a\validacao",
)

# 4) Mover extras com base no Excel
processor.mover_extras_do_excel(
    arquivo_excel_validacao=r"output\anexo_a\validacao\validacao_por_tipo.xlsx",
    raiz_output_telaspretas=r"output\anexo_a",
    mover=True,
)

# 5) Converter TXT -> PDF
processor.converter_pasta_txt_para_pdf(
    pasta_txt=r"output\anexo_a",
    pasta_pdf=r"output\anexo_a\pdfs",
    recursivo=True,
)

# 6) Merge global por contrato
processor.merge_global_por_contrato(
    pasta_base=r"output\anexo_a\pdfs",
    recursivo=True,
    nome_pasta_saida="merged",
)

# 7) Filtrar PDFs com base em CSV
processor.mover_pdfs_filtrados_por_csv(
    pasta_pdfs=r"output\anexo_a\pdfs\merged",
    csv_path=r"data\anexo_a\100 Kits_KPMG.csv",
    nome_pasta_saida="100 maiores",
    coluna_csv="Contrato",
    recursivo=True,
    copiar_ao_inves_de_mover=False,
)

# 8) Renomear PDFs com prefixo, se necessário
processor.renomear_arquivos_com_prefixo(
    pasta=r"output\anexo_a\pdfs\merged",
    prefixo="Anexo N_A",
    recursivo=False,
    apenas_pdfs=True,
)
```
