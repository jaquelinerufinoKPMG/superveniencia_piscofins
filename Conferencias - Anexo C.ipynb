```python
import os
import re
import shutil

def mover_arquivos_por_contrato(
    pasta_origem: str,
    pasta_destino: str,
    contratos: list[int | str],
):
    os.makedirs(pasta_destino, exist_ok=True)

    # contratos normalizados (sem zeros à esquerda)
    contratos_str = {str(int(c)) for c in contratos}

    movidos = []
    ignorados = []

    for nome_arquivo in os.listdir(pasta_origem):
        caminho_origem = os.path.join(pasta_origem, nome_arquivo)

        if not os.path.isfile(caminho_origem):
            continue

        # extrai todos os números do nome do arquivo
        numeros_no_nome = {str(int(n)) for n in re.findall(r"\d+", nome_arquivo)}

        if numeros_no_nome & contratos_str:
            caminho_destino = os.path.join(pasta_destino, nome_arquivo)
            shutil.move(caminho_origem, caminho_destino)
            movidos.append(nome_arquivo)
        else:
            ignorados.append(nome_arquivo)

    return movidos, ignorados
```

```python
import os
import re

def arquivos_sem_contrato(pasta: str, contratos: list[int | str]):
    """
    Retorna lista de arquivos da pasta cujo nome NÃO contém nenhum número
    presente na lista de contratos.
    """
    contratos_str = {str(int(c)) for c in contratos}

    nao_casados = []

    for nome_arquivo in os.listdir(pasta):
        caminho = os.path.join(pasta, nome_arquivo)
        if not os.path.isfile(caminho):
            continue

        numeros_no_nome = {str(int(n)) for n in re.findall(r"\d+", nome_arquivo)}

        if not (numeros_no_nome & contratos_str):
            nao_casados.append(nome_arquivo)

    return nao_casados
```
